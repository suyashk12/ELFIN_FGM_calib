```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.sparse import csc_matrix
from scipy.sparse.linalg import lsqr
import datetime as dt
from scipy.integrate import trapezoid, simpson
from scipy.optimize import curve_fit
```

```python
rootdir = 'scizone1_nooffset/'
```

```python
# Load in IGRF field and satellite attitude estimates in GEI
igrf_dat_dir = rootdir + 'elx_bt89_gei.txt'
att_dat_dir = rootdir + 'elx_att_gei.txt'
```

```python
igrf_data = np.loadtxt(igrf_dat_dir, dtype=str)
att_data = np.loadtxt(att_dat_dir, dtype=str)
```

```python
B_IGRF_GEI_x = np.float_(igrf_data[:,1])
B_IGRF_GEI_y = np.float_(igrf_data[:,2])
B_IGRF_GEI_z = np.float_(igrf_data[:,3])
```

```python
att_GEI_x = np.float_(att_data[:, 1])
att_GEI_y = np.float_(att_data[:, 2])
att_GEI_z = np.float_(att_data[:, 3])
```

```python
times = list(igrf_data[:,0])
times = [t.replace('/', ' ') for t in times]
data = pd.DataFrame(times, columns = ['Time (Full)'])
data['Time (Full)'] = pd.to_datetime(data['Time (Full)'])
data['Time (Full)'] = data['Time (Full)']-data['Time (Full)'][0]
data['ctime'] = data['Time (Full)'].dt.total_seconds()
data = data.drop('Time (Full)', axis=1)
ctime = np.array(data['ctime'])
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, B_IGRF_GEI_x, label='x')
ax.plot(ctime, B_IGRF_GEI_y, label='y')
ax.plot(ctime, B_IGRF_GEI_z, label='z')

ax.set_title('IGRF Field in GEI Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Field (nT)')
ax.legend()
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, att_GEI_x, label='x')
ax.plot(ctime, att_GEI_y, label='y')
ax.plot(ctime, att_GEI_z, label='z')

ax.set_title('Satellite Attitude in GEI Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Unit Vector Component')
ax.legend()
```

To define a despun spacecraft coordinate system (DSC = DMXL($t=0$)). To define this, we will need the unit vector $\hat{b} = B_{IGRF, GEI}(t=0) / ||B_{IGRF, GEI}(t=0)||$. We compute these unit vectors for all times below and plot their variation to get an understanding of how much the Earth's field in GEI coordinates varies over one science collection of ELFIN.

```python
B_hat_IGRF_GEI_x = B_IGRF_GEI_x/np.sqrt(B_IGRF_GEI_x**2 + B_IGRF_GEI_y**2 + B_IGRF_GEI_z**2)
B_hat_IGRF_GEI_y = B_IGRF_GEI_y/np.sqrt(B_IGRF_GEI_x**2 + B_IGRF_GEI_y**2 + B_IGRF_GEI_z**2)
B_hat_IGRF_GEI_z = B_IGRF_GEI_z/np.sqrt(B_IGRF_GEI_x**2 + B_IGRF_GEI_y**2 + B_IGRF_GEI_z**2)
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, B_hat_IGRF_GEI_x, label='x')
ax.plot(ctime, B_hat_IGRF_GEI_y, label='y')
ax.plot(ctime, B_hat_IGRF_GEI_z, label='z')

ax.set_title('IGRF Field Directions in GEI Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Unit Vector Component')
ax.legend()
```

We also have $\hat{u}$, the attitude vector of the spacecraft. While we have a guess for the time variation of this quantity over the course of this particular science collection, the attitude is almost constant throughout. We are creating a synthetic dataset here, we will use the average value of the attitude vector over the collection as the "true" attitude.

```python
u_hat = np.array([np.mean(att_GEI_x), np.mean(att_GEI_y), np.mean(att_GEI_z)])
b_hat = np.array([B_hat_IGRF_GEI_x[0], B_hat_IGRF_GEI_y[0], B_hat_IGRF_GEI_z[0]])
```

```python
u_hat
```

The despun spacecraft coordinate system (DSC) is defined as the trio $[(\hat{b} \times \hat{u}), \hat{u} \times (\hat{b} \times \hat{u}) ,\hat{u}]$, of course with each vector normalized. This transformation would help us move from DSC to GEI.

```python
DSC_2_GEI = np.zeros((3, 3))

DSC_2_GEI[:,0] = np.cross(b_hat, u_hat)
DSC_2_GEI[:,1] = np.cross(u_hat, np.cross(b_hat, u_hat))
DSC_2_GEI[:,2] = u_hat

DSC_2_GEI[:,0] /= np.linalg.norm(DSC_2_GEI[:,0])
DSC_2_GEI[:,1] /= np.linalg.norm(DSC_2_GEI[:,1])
DSC_2_GEI[:,2] /= np.linalg.norm(DSC_2_GEI[:,2])
```

```python
GEI_2_DSC = np.linalg.inv(DSC_2_GEI)
```

We can transform the IGRF field from GEI coordinates to DSC coordinates.

```python
B_IGRF_DSC_x = GEI_2_DSC[0,0]*B_IGRF_GEI_x + GEI_2_DSC[0,1]*B_IGRF_GEI_y + GEI_2_DSC[0,2]*B_IGRF_GEI_z
B_IGRF_DSC_y = GEI_2_DSC[1,0]*B_IGRF_GEI_x + GEI_2_DSC[1,1]*B_IGRF_GEI_y + GEI_2_DSC[1,2]*B_IGRF_GEI_z
B_IGRF_DSC_z = GEI_2_DSC[2,0]*B_IGRF_GEI_x + GEI_2_DSC[2,1]*B_IGRF_GEI_y + GEI_2_DSC[2,2]*B_IGRF_GEI_z
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, B_IGRF_DSC_x, label='x')
ax.plot(ctime, B_IGRF_DSC_y, label='y')
ax.plot(ctime, B_IGRF_DSC_z, label='z')

ax.set_title('IGRF Field in DSC Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Field (nT)')
ax.legend()
```

Next up, we spin up the IGRF field from DSC to respun spacecraft coordinates (RSC). Again, we specify a "true" period of 2.84s and a zero-phase of $\pi$.

```python
T = 2.84
w = 2*np.pi/T
theta0 = np.pi

theta = w*ctime + theta0
```

```python
B_IGRF_RSC_x = np.cos(-theta)*B_IGRF_DSC_x - np.sin(-theta)*B_IGRF_DSC_y
B_IGRF_RSC_y = np.sin(-theta)*B_IGRF_DSC_x + np.cos(-theta)*B_IGRF_DSC_y
B_IGRF_RSC_z = B_IGRF_DSC_z
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, B_IGRF_RSC_x, label='x')
ax.plot(ctime, B_IGRF_RSC_y, label='y')
ax.plot(ctime, B_IGRF_RSC_z, label='z')

ax.set_title('IGRF Field in RSC Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Field (nT)')
ax.set_xlim(0, 30)
ax.legend()
```

Next, we use the transformation -

$$\Gamma = \begin{pmatrix}
\cos(f) & \sin(f) & 0\\
0 & 0 & 1\\
\sin(f) & -\cos(f) & 0
\end{pmatrix}$$

To transform from FGM to RSC coordinates.

```python
f = 44*np.pi/180
```

```python
FGM_2_RSC = np.array([[np.cos(f),np.sin(f),0],[0,0,1],[np.sin(f),-np.cos(f),0]])
```

```python
RSC_2_FGM = np.linalg.inv(FGM_2_RSC)
```

```python
B_IGRF_FGM_x = RSC_2_FGM[0,0]*B_IGRF_RSC_x + RSC_2_FGM[0,1]*B_IGRF_RSC_y + RSC_2_FGM[0,2]*B_IGRF_RSC_z
B_IGRF_FGM_y = RSC_2_FGM[1,0]*B_IGRF_RSC_x + RSC_2_FGM[1,1]*B_IGRF_RSC_y + RSC_2_FGM[1,2]*B_IGRF_RSC_z
B_IGRF_FGM_z = RSC_2_FGM[2,0]*B_IGRF_RSC_x + RSC_2_FGM[2,1]*B_IGRF_RSC_y + RSC_2_FGM[2,2]*B_IGRF_RSC_z
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, B_IGRF_FGM_x, label='x')
ax.plot(ctime, B_IGRF_FGM_y, label='y')
ax.plot(ctime, B_IGRF_FGM_z, label='z')

ax.set_title('IGRF Field in FGM Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Field (nT)')
ax.set_xlim(0, 10)
ax.legend()
```

We apply the offsets and orthogonality matrix to the IGRF fields in FGM coordinates - this creates the corrupted sensor data.

```python
orth = np.array([[1.01, 0.05, 0.04], [0.02, 1.0, 0.03], [0.006, 0.006, 1.0]])
offsets = np.array([3, -4, -5])
#offsets = np.array([100,100,100])

#orth = np.array([[1,0,0],[0,1,0],[0,0,1]])
#offsets = np.array([0,0,0])
```

```python
np.round(orth, 2)
```

```python
np.round(offsets, 2)
```

```python
B_raw_FGM_x = orth[0,0]*B_IGRF_FGM_x + orth[0,1]*B_IGRF_FGM_y + orth[0,2]*B_IGRF_FGM_z + offsets[0]
B_raw_FGM_y = orth[1,0]*B_IGRF_FGM_x + orth[1,1]*B_IGRF_FGM_y + orth[1,2]*B_IGRF_FGM_z + offsets[1]
B_raw_FGM_z = orth[2,0]*B_IGRF_FGM_x + orth[2,1]*B_IGRF_FGM_y + orth[2,2]*B_IGRF_FGM_z + offsets[2]
```

```python
fig, ax = plt.subplots(1, figsize = (15, 7))

ax.plot(ctime, B_raw_FGM_x, label='x')
ax.plot(ctime, B_raw_FGM_y, label='y')
ax.plot(ctime, B_raw_FGM_z, label='z')

ax.set_title('Uncalibrated Sensor Field in FGM Coordinates')
ax.set_xlabel('Relative Time (s)')
ax.set_ylabel('Field (nT)')
ax.set_xlim(0, 10)
ax.legend()
```
