First, we define some utilities for curve fitting

```python
def linear_fit(x, m, c):
    return m*x + c
```

```python
def quad_fit(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
def cube_fit(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
def sine_fit(x, alpha, A, w, p, k):
    return alpha*A*np.sin(w*x+p) + k
```

```python
def cosine_fit(x, alpha, A, w, p, k):
    return alpha*A*np.cos(w*x+p) + k
```

This function isolates retains the positive elements of an array $L$ and marks the non-positive elements as <code>NaN</code>. This will be helpful later for obtaining rotation angles of the satellite via numerically integrating the synodic angular velocity.

```python
def mask_neg(L):
    
    L_mask = np.zeros(len(L))
    
    for i in range(len(L)):
        
        if(L[i]>0):
            L_mask[i] = L[i]
        else:
            L_mask[i] = np.nan
            
    return L_mask
```

Generally, we will be using the cubic function for interpolation throughout this notebook, so we'll just globally declare a fitting function <code>func</code> to be invoked later. You may change the values of <code>func</code> to either <code>quad_fit</code> or <code>linear_fit</code> to understand the impact of using those fitting functions.

```python
func = cube_fit
```

We define some filtering functions below.

Moving average produces a smoothed out version of a <code>signal</code> sampled at instances <code>time</code> by replacing <code>signal[i]</code> with the fitted version (in this case quadratic, as specified by <code>quad_fit</code>) using samples within a period of <code>T</code> around <code>time[i]</code>. This will be used for performing a low-pass filter, which is the compliment of a high-pass filter that is run on the calibrated signal in DMXL coordinates to discover EMIC waves and Field Aligned Currents.

```python
def moving_average(time, signal, T=10, func=quad_fit):
    
    signal_avg = np.zeros(len(signal))
    
    for i in range(len(signal)):
    
        if(time[i]-time[0]<T/2):
            low_lim = time[i]
            high_lim = time[i]+T
        elif(time[-1]-time[i]<T/2):
            low_lim = time[i]-T
            high_lim = time[i]
        else:   
            low_lim = time[i]-T/2
            high_lim = time[i]+T/2

        idx = (time>=low_lim)&(time<=high_lim)

        fit_opt, fit_covar = curve_fit(func, time[idx], signal[idx])

        signal_avg[i] = func(time[i], *fit_opt)
    
    return signal_avg 
```

The <code>running_spline</code> is indeed simiilar to <code>moving_average</code>, except that the sampling times <code>sample_time</code> where the fitted function is evaluated can be different from the <code>time</code> where the actual <code>sample_signal</code> is provided. This will be used for performing a running spline over points of the synodic angular velocity later.

**Note** - The <code>moving_average</code> function can be wrapped into <code>running_spline</code> by setting <code>time = sample_time</code>.

```python
def running_spline(time, sample_time, sample_signal, T=10, func=cube_fit):
    
    signal_fit = np.zeros(len(time))
    
    for i in range(len(time)):
    
        if(time[i]-time[0]<T/2):
            low_lim = time[0]
            high_lim = time[0]+T
        elif(time[-1]-time[i]<T/2):
            low_lim = time[-1]-T
            high_lim = time[-1]
        else:   
            low_lim = time[i]-T/2
            high_lim = time[i]+T/2

        idx = (time>=low_lim)&(time<=high_lim)
        idx_sample = (sample_time>=low_lim)&(sample_time<=high_lim)

        fit_opt, fit_covar = curve_fit(func, sample_time[idx_sample], sample_signal[idx_sample])

        signal_fit[i] = func(time[i], *fit_opt)
    
    return signal_fit
```

The structure of the <code>running_filter</code> is very similar to <code>moving_average</code> and <code>running_spline</code>, except now we compute both a running average along with a running standard deviation, and retain sample values that fall within some fraction (specified by <code>eps</code>) of a standard deviation of the median. This will be used for discarding unusual spin-periods that occur as a result of gappy data and therefore interfere with the determination of synodic angular velocities.

```python
def running_filter(time, signal, eps=1, T=20):
    
    filter_idx = []
    
    for i in range(len(time)):
    
        if(time[i]-time[0]<T/2):
            low_lim = time[0]
            high_lim = time[0]+T
        elif(time[-1]-time[i]<T/2):
            low_lim = time[-1]-T
            high_lim = time[-1]
        else:   
            low_lim = time[i]-T/2
            high_lim = time[i]+T/2
            
        idx = (time>=low_lim)&(time<=high_lim)
        
        med = np.median(signal[idx])
        std = np.std(signal[idx])
        
        if (signal[i]>=med-eps*std and signal[i]<=med+eps*std):
            filter_idx.append(i)
            
    return np.array(filter_idx)
```

Onto estimating the synodic angular velocities! This is done over three stages. First, we obtain the times at which $B_{S3}$ is maximized, i.e. descending zero-crossings for the derivative of $B_{S3}$. Note that using the derivative has the benefit of avoiding any deviations in maxima that sensor offsets may introduce; but some corruption from the orthogonality matrix will still occur.

```python
# Stage one - crossing times

d_B_S3_corr = np.gradient(B_S3_corr)/np.gradient(ctime)

cross_times_d_B_S3_corr_1 = []

for i in range(len(ctime)-1):
    if(d_B_S3_corr[i]*d_B_S3_corr[i+1]<0 and d_B_S3_corr[i]>0):
        y1 = d_B_S3_corr[i]
        y2 = d_B_S3_corr[i+1]
        x1 = ctime[i]
        x2 = ctime[i+1]
        cross_times_d_B_S3_corr_1.append((y2*x1 - y1*x2)/(y2 - y1))
        
# List of crossing times
cross_times_d_B_S3_corr_1 = np.array(cross_times_d_B_S3_corr_1)

# List of middle points of crossing times, helpful for interpolation purposes earlier
cross_times_d_B_S3_corr_1_mids = .5*(cross_times_d_B_S3_corr_1[1:]+cross_times_d_B_S3_corr_1[:-1])

# Spin-periods computed as difference between consecutive zero-crossings
T_spins_d_corr_1 = cross_times_d_B_S3_corr_1[1:]-cross_times_d_B_S3_corr_1[:-1]

# Corresponding angular velocities
w_syn_d_corr_1 = 2*np.pi/T_spins_d_corr_1
```

We can refine our zero-crossing estimates further! Instead of noting where the signal itself (read derivative of $B_{S3}$) goes from positive to negative and linearly interpolating between those points, we can estimate the signal within one period (esitmated from stage 1) around the zero-crossings (again estimated from stage 1) using a sinusoid, and can then see where the phase (which would vary linearly in that short period) goes from positive to negative - this interpolation would give us a crossing time much more closer to reality!

```python
# Stage two - refine crossing times

# For noting phases for determining zero-crossings

phase_corr = np.zeros(len(ctime))

for i in range(len(cross_times_d_B_S3_corr_1_select)):
    
    # Zero-crossing from stage 1
    t0 = cross_times_d_B_S3_corr_1_select[i]
    
    # Isolate a period around the zero-crossing
    idx = ((ctime-t0)>=-T_spins_d_pad_corr_1_select[i]/2) & ((ctime-t0)<=T_spins_d_pad_corr_1_select[i]/2)
    
    # Use the arcsine function to get phase angle around the zero-crossing
    phase_corr[idx] = np.arcsin(d_B_S3_corr[idx]/np.max(np.abs(d_B_S3_corr[idx])))
    
# Record zero crossings as locations of the phase passing over from positive to negative
cross_times_d_B_S3_corr_2 = []

for i in range(len(ctime)-1):
    if(phase_corr[i]*phase_corr[i+1]<0 and phase_corr[i]>0):
        y1 = phase_corr[i]
        y2 = phase_corr[i+1]
        x1 = ctime[i]
        x2 = ctime[i+1]
        cross_times_d_B_S3_corr_2.append((y2*x1 - y1*x2)/(y2 - y1))
        
# Obtaining synodic angular velocity samples - similar to stage 1
cross_times_d_B_S3_corr_2 = np.array(cross_times_d_B_S3_corr_2)
cross_times_d_B_S3_corr_2_mids = .5*(cross_times_d_B_S3_corr_2[1:]+cross_times_d_B_S3_corr_2[:-1])
T_spins_d_corr_2 = cross_times_d_B_S3_corr_2[1:]-cross_times_d_B_S3_corr_2[:-1]
w_syn_d_corr_2 = 2*np.pi/T_spins_d_corr_2
```

We can go up one last step in the ladder of sophistication when determining the zero crossings. We can isolate <code>N_spins_fit</code> number of spins around the zero-crossings (this time from stage 2), and then simply fit a sine curve to it, also fitting for the phase offset. Using the optimal phase offset, we can infer the perturbation to the zero-crossing times that were computed in stage 2. Check the comments alongside the code for more details!

```python
N_spins_fit = 4

# For detecting peaks by fitting B_S3 itself instead of 
# fitting its derivative and computing zero-crossings
peak_detect = False
```

```python
# Stage three - refine crossing times again

cross_times_d_B_S3_corr_3 = []
w_syn_d_corr_3 = []

for i in range(len(cross_times_d_B_S3_corr_2_select)):

    # Get the crossing time, synodic period, and angular velocity from stage 2
    t0 = cross_times_d_B_S3_corr_2_select[i]
    T_syn = T_spins_d_pad_corr_2_select[i]
    w_syn = 2*np.pi/T_syn
    
    # The total time over which the fit will be computed
    T_avg = N_spins_fit*T_syn
    
    # Dealing with edge cases around the beginning and the end of time
    # Similar idea to moving_average, running_spline, and running-filter
    if(t0-ctime[0]<T_avg/2):
        low_lim = ctime[0]
        high_lim = ctime[0]+T_avg
    elif(ctime[-1]-t0<T_avg/2):
        low_lim = t0-T_avg
        high_lim = t0
    else:   
        low_lim = t0-T_avg/2
        high_lim = t0+T_avg/2

    # Get time instances within N_spins_fit
    idx = (ctime>=low_lim) & (ctime<=high_lim)

    # Initialize time for this segment at the zero-crossing
    ctime_slice = ctime[idx]-t0

    # Slice the signal itself
    
    # In case you are trying to find the maxima of B_S3 directly
    if(peak_detect == True):
        signal_slice = B_S3_corr[idx]
        spin_func = lambda x, A, w, p: cosine_fit(x, -1, A, w, p, 0)

    # In case you are trying to go the derivative/ zero-crossing route
    else:        
        signal_slice = d_B_S3_corr[idx]
        spin_func = lambda x, A, w, p: sine_fit(x, 1, A, w, p, 0)

    # Fit the curve you want to work with!
    # Good initial guesses p0 really help
    spin_opt, spin_covar = curve_fit(spin_func, 
                                     ctime_slice, 
                                     signal_slice, 
                                     p0=[np.max(np.abs(signal_slice-np.mean(signal_slice))), 
                                         w_syn, 0])

    # Using the zero-phase and the angular velocity, computing the deviation to the crossing time
    delta_t0 = -spin_opt[2]/spin_opt[1]

    # Contextualize the computing perturbation in terms of the relative time for the science zone
    cross_times_d_B_S3_corr_3.append(t0+delta_t0)
    # Also save the fitted value for the angular velocity
    w_syn_d_corr_3.append(spin_opt[1])

cross_times_d_B_S3_corr_3 = np.array(cross_times_d_B_S3_corr_3)
w_syn_d_corr_3 = np.array(w_syn_d_corr_3)
```

Note that in this case, the angular velocity samples are generated directly at the zero-crossing times, so no need to pad like in stage 1 and 2!

Next, we determine the rotation angles from the synodic angular velocity at all times when the FGM readings are sampled. The simplest recipe is to use the first zero-crossing as the reference and perform a numerical integral from the zero-crossing to all sampled times (there is a small correction that needs to be made considering the fact that the first zero-crossing may not be among the sampled times). 

But considering the significance of minuscule variations in the zero-crossing times, it might be a better idea to do numerical integration for each time relative to its "closest" zero-crossing (<code>relative_integrate = True</code>).

What "closest" means can have two interpretations. A simpler interpretation, again, is the closest zero-crossing to the left, i.e. before a given time, except of course in the case of times before the first zero-crossing, when we need to use the first zero-crossing. But we can also use the closest zero-crossing in an absolute sense (i.e. using the closest crossing time irrespective of it being before or after that given time, <code>bidirectional_integrate = True</code>).

Based on the assignments of each of the below booleans, there can be several subcases which have been dealt with carefully in the code that follows. Read comments for more details!

```python
relative_integrate = True
bidirectional_integrate = True
```

The code below obtains the indices of the times that fall right before the crossing time(s). This is important to do because when performing numerical integration over the synodic angular velocities sampled at the FGM reading times will require a correction for the fact that the sample times may not include the crossing times. Thereafter, we also properly compute the angular velocities at the crossing times. If the zero-crossing method is 3, doing so should not be necessary.

```python
if (relative_integrate == True):
    
    # Use multiple reference points for integration

    idx0_corrs = np.array([np.where(ctime<=t0_corr)[0][-1] for t0_corr in cross_times_d_B_S3_corr])

    if (fit_running_spline == True):

        if(zero_crossing_method == 1 or zero_crossing_method == 2):
            w_t0_corrs = running_spline(cross_times_d_B_S3_corr, cross_times_d_B_S3_corr_mids, w_syn_d_corr, T=30)
        else:
            w_t0_corrs = running_spline(cross_times_d_B_S3_corr, cross_times_d_B_S3_corr, w_syn_d_corr, T=50)

    else:
        if(zero_crossing_method == 1 or zero_crossing_method == 2):
            w_t0_corrs = func(cross_times_d_B_S3_corr, *curve_fit(func, cross_times_d_B_S3_corr_mids, w_syn_d_corr)[0])
        else:
            w_t0_corrs = func(cross_times_d_B_S3_corr, *curve_fit(func, cross_times_d_B_S3_corr, w_syn_d_corr)[0])
            
else:
    
    # Use just one reference point for integration
    
    t0_corr = cross_times_d_B_S3_corr[0]
    idx0_corr = np.where(ctime<=t0_corr)[0][-1]

    if (fit_running_spline == True):

        if(zero_crossing_method == 1 or zero_crossing_method == 2):
            w_t0_corr = running_spline([t0_corr], cross_times_d_B_S3_corr_mids, w_syn_d_corr, T=30)[0]
        else:
            w_t0_corr = running_spline([t0_corr], cross_times_d_B_S3_corr, w_syn_d_corr, T=50)[0]

    else:
        if(zero_crossing_method == 1 or zero_crossing_method == 2):
            w_t0_corr = func(t0_corr, *curve_fit(func, cross_times_d_B_S3_corr_mids, w_syn_d_corr)[0])
        else:
            w_t0_corr = func(t0_corr, *curve_fit(func, cross_times_d_B_S3_corr, w_syn_d_corr)[0])
```

Next, we determine the rotation angles, case by case depending on choices for <code>relative_integrate</code> and <code>bidirectional_integrate</code>.

```python
phi_corr = np.zeros(len(ctime))

for i in range(len(phi_corr)):
    
    if (relative_integrate == True):
        
        if(bidirectional_integrate == True):
            
            # In this case, find the closest zero-crossings in absolute sense

            t0_corr_idx = np.argmin(np.abs(ctime[i]-cross_times_d_B_S3_corr))

        else:
            
            # In this case, find the closest zero crossing that comes before a given time
            # Except for times before the zero crossing, then use the first zero crossing
            
            if(np.sum((ctime[i]-cross_times_d_B_S3_corr)>0) == 0):
                t0_corr_idx = 0
            else:        
                t0_corr_idx = np.nanargmin(mask_neg(ctime[i]-cross_times_d_B_S3_corr))
        
        t0_corr = cross_times_d_B_S3_corr[t0_corr_idx]
        idx0_corr = idx0_corrs[t0_corr_idx]
        w_t0_corr = w_t0_corrs[t0_corr_idx]

    # Do the integral based on the relative position of the time and the relevant zero-crossing
    
    if (i<idx0_corr):
        phi_corr[i] = -simpson(w_syn_corr[i:idx0_corr+1], x=ctime[i:idx0_corr+1], dx=delta_t)
    elif (i>idx0_corr):
        phi_corr[i] = simpson(w_syn_corr[idx0_corr:i+1], x=ctime[idx0_corr:i+1], dx=delta_t)
    else:
        phi_corr[i] = 0
        
    # Correct for the zero-crossing not being in the time array  
    
    phi_corr[i] -= .5*(w_t0_corr+w_syn_corr[idx0_corr])*(t0_corr-ctime[idx0_corr])
```

Get the calibration parameters (sensor angles, gains, and offsets) using normal equations and linear least squares inversion

```python
n = len(B_S1_corr)
```

```python
b = np.concatenate((B_S1_corr, B_S2_corr, B_S3_corr))
```

```python
A = np.zeros((3*n, 12))
```

```python
A[0:n,0] = B_S1_IGRF
A[0:n,1] = B_S2_IGRF
A[0:n,2] = B_S3_IGRF
A[0:n,3] = np.ones(n)
A[n:2*n,4] = B_S1_IGRF
A[n:2*n,5] = B_S2_IGRF
A[n:2*n,6] = B_S3_IGRF
A[n:2*n,7] = np.ones(n)
A[2*n:3*n,8] = B_S1_IGRF
A[2*n:3*n,9] = B_S2_IGRF
A[2*n:3*n,10] = B_S3_IGRF
A[2*n:3*n,11] = np.ones(n)
```

```python
A = csc_matrix(A)
```

Solve the least squares problem

```python
x = lsqr(A, b, atol=1e-10, btol=1e-10)[0]
```

Isolate the orthogonality matrix and the offsets from the solution

```python
orth = np.array([[x[0],x[1],x[2]], [x[4], x[5], x[6]], [x[8], x[9], x[10]]])
offsets = np.array([x[3], x[7], x[11]])
#offsets = np.array([ -815.45,    51.93, -1083.54])
```

```python
np.round(orth, 2)
```

```python
np.round(offsets, 2)
```

```python
calib = np.linalg.inv(orth)
np.round(calib, 2)
```

Apply the calibration to the corrupted sensor data

```python
B_S1_calib = calib[0, 0]*(B_S1_corr-offsets[0]) + calib[0, 1]*(B_S2_corr-offsets[1]) + calib[0, 2]*(B_S3_corr-offsets[2])
B_S2_calib = calib[1, 0]*(B_S1_corr-offsets[0]) + calib[1, 1]*(B_S2_corr-offsets[1]) + calib[1, 2]*(B_S3_corr-offsets[2])
B_S3_calib = calib[2, 0]*(B_S1_corr-offsets[0]) + calib[2, 1]*(B_S2_corr-offsets[1]) + calib[2, 2]*(B_S3_corr-offsets[2])
```

Applicable for real ELFIN data - we can try to detect field aligned currents more properly. Begin with obtaining the spin-periods at each zero-crossing, depending on your chosen method.

```python
if (zero_crossing_method == 1):
    
    T_spins_d_calib = T_spins_d_pad_calib_1_select
    
elif (zero_crossing_method == 2):
    
    T_spins_d_calib = T_spins_d_pad_calib_2_select
    
else:
    
    T_spins_d_calib = 2*np.pi/w_syn_d_calib_3_select
```

Construct the FACs in each direction (in DMXL coordinates) by fitting a constant offset + sine curve on a spin-period's worth of readings around the zero-crossings. We just retain phase constant offset recorded from the fitting to report the FACs.

```python
FAC_smooth_x = np.zeros(len(cross_times_d_B_S3_calib))
FAC_smooth_y = np.zeros(len(cross_times_d_B_S3_calib))
FAC_smooth_z = np.zeros(len(cross_times_d_B_S3_calib))
```

```python
for i in range(0, len(cross_times_d_B_S3_calib)):
    
    t0 = cross_times_d_B_S3_calib[i]
    T_syn = T_spins_d_calib[i]
    w_syn = 2*np.pi/T_syn
    
    idx = ((ctime-t0)>=-.5*T_syn) & ((ctime-t0)<=.5*T_syn)
    
    ctime_slice = ctime[idx]
    
    B_x_DMXL_resid_slice = (B_x_DMXL_calib-B_x_DMXL_true)[idx]
    B_y_DMXL_resid_slice = (B_y_DMXL_calib-B_y_DMXL_true)[idx]
    B_z_DMXL_resid_slice = (B_z_DMXL_calib-B_z_DMXL_true)[idx]
    
    FAC_func = lambda x, A, k: sine_fit(x, 1, A, w_syn, -w_syn*t0, k)
    
    FAC_smooth_x[i] = curve_fit(FAC_func, ctime_slice, B_x_DMXL_resid_slice)[0][1]
    FAC_smooth_y[i] = curve_fit(FAC_func, ctime_slice, B_y_DMXL_resid_slice)[0][1]
    FAC_smooth_z[i] = curve_fit(FAC_func, ctime_slice, B_z_DMXL_resid_slice)[0][1]
```

Here are the beautiful FACs - this should form the data product!

```python
fig, ax = plt.subplots(1, figsize=(15,7))

ax.plot(cross_times_d_B_S3_calib, FAC_smooth_x, label='x', alpha=.5)
ax.plot(cross_times_d_B_S3_calib, FAC_smooth_y, label='y', alpha=.5)
ax.plot(cross_times_d_B_S3_calib, FAC_smooth_z, label='z', alpha=.5)

ax.set_title('Spin-period LMFIT IGRF Residuals (in DMXL)')
ax.set_xlabel('Relative Time (seconds)')
ax.set_ylabel('Field (nT)')
#ax.set_xlim(100,250)
#ax.set_ylim(-7,7)
ax.legend()
```
